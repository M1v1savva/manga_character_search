```python
from tensorflow.keras.applications.vgg19  import VGG19
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input
from tensorflow.keras.models import Model
import numpy as np
from matplotlib.pyplot import imshow, subplots
from sklearn.neighbors import KNeighborsClassifier
from os import walk
```

```python
model = VGG19(weights='imagenet', include_top=False, pooling='avg')
```

```python
def get_features(img_path, subs = None):
    img = image.load_img(img_path, target_size=(224, 224))
    if subs != None:
        subs.imshow(img)
        
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    features = model.predict(x)
    return features
```

```python
l_dir = 'characters/cropped/Levi Ackerman'
l_files = next(walk(l_dir), (None, None, []))[2]
print(l_files)
```

```python
fig, subs = subplots(1, 4, figsize=(20, 20))
l_train = [None] * 4

for i in range(4):
    l_train[i] = get_features(l_dir + '/' + l_files[i], subs[i])
```

```python
m_dir = 'characters/cropped/Mikasa Ackerman'
m_files = next(walk(m_dir), (None, None, []))[2]
print(m_files)
```

```python
fig, subs = subplots(1, 4, figsize=(20, 20))
m_train = [None] * 4

for i in range(4):
    m_train[i] = get_features(m_dir + '/' + m_files[i], subs[i])
```

```python
fig, subs = subplots(1, 2, figsize=(7, 7))
lt = get_features('characters/cropped/test/l.jpg', subs[0])
mt = get_features('characters/cropped/test/m.jpg', subs[1])
```

```python
X_train = l_train + m_train
y_train = ['levi'] * 4 + ['mikasa'] * 4
X_test = [lt, mt]

X_train = [item.flatten() for item in X_train]
X_test = [item.flatten() for item in X_test]
```

```python
def run_knn(n_neighbors):
    neigh = KNeighborsClassifier(n_neighbors=n_neighbors)
    neigh.fit(X_train, y_train)
    y_test = neigh.predict(X_test)
    print(y_test)
    prob = neigh.predict_proba(X_test)
    print(prob)
```

```python
for n_neighbors in range(3, 6, 2):
    run_knn(n_neighbors)
```
